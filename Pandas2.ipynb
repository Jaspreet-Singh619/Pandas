```python
import pandas as pd

index_levels = ['Continent', 'Country']
data = [['North America', 'Canada'], ['North America', 'USA'], ['Europe', 'France']]
multi_index = pd.MultiIndex.from_tuples(data, names=index_levels)

df = pd.DataFrame({'data': [10, 20, 30]}, index=multi_index)
print(df)
```
